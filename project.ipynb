```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def fun(x):
    return (0.1*x*x*x - 2*x*x + 10*math.sqrt(x))


def mkv(h, x1, e):
    xm = x1
    f0 = fun(x1 - h)
    f1 = fun(x1)
    f2 = fun(x1 + h)

    c = (f0 - 2*f1 + f2) / (2*h*h)
    b = ((-1)*f0*(2*x1 + h) + 4*f1*x1 - f2*(2*x1-h)) / (2*h*h)

    xm = (-1) * b / (2 * c)

    while ((xm - x1) < e):
        x1 = xm
        f0 = fun(x1 - h)
        f1 = fun(x1)
        f2 = fun(x1 + h)

        c = (f0 - 2*f1 + f2) / (2*h*h)
        b = ((-1)*f0*(2*x1 + h) + 4*f1*x1 - f2*(2*x1-h)) / (2*h*h)

        xm = (-1) * b / (2 * c)
    return [xm, fun(xm)]
```

```python
#a
x = np.linspace(1, 2, 100)
y1 = x ** mkv(1, 1, 0.1)[0]
y2 = x ** mkv(1, 1, 0.1)[1]


fig, ax = plt.subplots()                        # будет 1 график, на нем:
ax.plot(x, y1, color="blue", label="y1")
ax.plot(x, y2, color="red", label="y2")
ax.set_xlabel("x")
# ax.set_ylabel("y")                              # подпись у вертикальной оси y
ax.legend()                                     # показывать условные обозначения

plt.show()                                      # показать рисунок
```
